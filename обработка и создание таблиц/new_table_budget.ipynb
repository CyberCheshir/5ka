```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime, time 
```

```python
df_march = pd.read_excel('data/3.budget.xlsx',sheet_name='март')
df_april = pd.read_excel('data/3.budget.xlsx', sheet_name='апрель')
df_may = pd.read_excel('data/3.budget.xlsx',sheet_name='май')

book_budget_march = dict(zip(df_march['Завод'],df_march['Сумма']))
book_budget_april = dict(zip(df_april['Завод'], df_april['Сумма']))
book_budget_may = dict(zip(df_may['Завод'],df_may['Сумма']))

df_id_market = pd.read_excel('new_data/id_market_edit.xlsx')
df_march
```

```python
sap = df_id_market['SAP']
num_market = df_id_market['№ магазина']
book_mark_to_sap = dict(zip(num_market,sap))
book_sap_to_mark = dict(zip(sap,num_market))
temp_df = pd.DataFrame()
```

```python
def if_in_market_budget(sap_arr):
    arr = []
    for i in sap_arr:
        if book_sap_to_mark.get(i)!= None:
            arr.append(book_sap_to_mark.get(i))
    return arr

# Пересеченеи множеств месяцев. Мы должны прогнозировать те месяца, которые имеют бюджеты по 3 месяцам  (без указания пустого бюджета)
s1 = set(if_in_market_budget(df_march['Завод']))
s2 = set(if_in_market_budget(df_april['Завод']))
s3 = set(if_in_market_budget(df_may['Завод']))

general_array_market = s1.intersection(s2).intersection(s3)
print(len(set(df_id_market['SAP']).intersection(general_array_market)))
```

```python

temp_df['Магазин'] = if_in_market_budget(general_array_market)
temp_df['SAP'] = [book_mark_to_sap[i] for i in temp_df['Магазин']]
temp_df[:5]
```

```python
temp_df['Расчет бюджета март'] = [book_budget_march[i] for i in temp_df['SAP']]
temp_df['Расчет бюджета апрель'] = [book_budget_april[i] for i in temp_df['SAP']]
temp_df['Расчет бюджета май'] = [book_budget_may[i] for i in temp_df['SAP']]
```
